```python
#Задача 0

inputs = []

while True:
    s = input()
    if '0' in s.split():
        break
    inputs.append(s)

numbers = [
    int(x)
    for s in inputs
    for x in s.split()
    if x.isdigit() and int(x) > 0
]

print(numbers)
print(len(numbers))
```

```python
#Задача 1 (Базовое преобразование)

celsius = [0, 15, 25, 30, 40, 100]
fahrenheit = [c * 9/5 + 32 for c in celsius]

print(f'Градусы по Цельсию: {celsius}')
print(f'Градусы Фаренгейт: {fahrenheit}')

grades = [45, 85, 92, 33, 67, 78, 90, 55, 29, 88]
passed = [grade for grade in grades if grade >= 60]

print(f'Оценки сдавших: {passed}')
```

```python
#Задача 3 (Работа со строками и вложенными циклами)

products = ['футболка', 'кружка', 'блокнот']
colors = ['красный', 'синий', 'зеленый']

combinations = [f'{product} - {color}' for product in products for color in colors]

print(*combinations, sep='\n')
```

```python
#Задача 4 (Сложная логика с тернарным оператором)

employees = [('Иван', 45), ('Мария', 92), ('Петр', 33), ('Анна', 67)]

performance = [
    'Отлично' if percent >= 90 else
    'Хорошо' if percent >= 60 else
    'Требует улучшения'
    for name, percent in employees
]

for i in range(len(employees)):
    name, percent = employees[i]
    print(f'{name}: {percent}% - {performance[i]}')
```

```python
#Задача 5 (Комплексная работа с данными)

sales = [
    {'name': 'Алексей', 'sales': 150, 'returns': 10},
    {'name': 'Ольга', 'sales': 200, 'returns': 5},
    {'name': 'Дмитрий', 'sales': 80, 'returns': 25},
    {'name': 'Елена', 'sales': 300, 'returns': 8}
]

top_managers = [
    manager['name']
    for manager in sales
    if manager['sales'] - manager['returns'] >= 150
    and manager['returns'] / manager['sales'] * 100 < 10
]

print('Топ-менеджеры:', top_managers)
```
